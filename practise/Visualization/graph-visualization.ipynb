```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(-20, 20, 1000)  # X-axis data
#y = [2, 4, 6, 8, 10]
y =  np.tanh(x) # Y-axis data, representing a linear relationship y = 2x
print(x,y)

# Create the plot
plt.plot(x, y, marker='o')  # 'o' creates circular markers at the data points

# Add titles and labels
plt.title('Linear Graph Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
